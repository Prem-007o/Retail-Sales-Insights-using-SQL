```python
import pandas as pd
df = pd.read_csv('orders.csv',na_values=['Not Available','unknown'])
```

```python
#df.rename(columns={'Order Id':'order_id','Order Date':'order_date','Ship Mode':'ship_mode'}) 
# there is more pro way
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df['discount']=df['list_price']*df['discount_percent']/100
df.head(10)
```

```python
df['sale_price']=df['list_price']-df['discount']
df.head(5)
```

```python
df['profit'] = df['sale_price'] -df['cost_price']
#if u didn't put equals to what column are u going to add .. it will create a temp colum..it wont change anything.... we have to assign the name with opj
#we use this for not to mess up anything .. like if we alredy assign the value.. what if it could g wrong.. so for double checking purose we need to see the temp data first .. then we can assign the coulm value..
df.head(5)
```

```python
df.drop(columns=['list_price','discount_percent','cost_price'],inplace=True)
#the inplace confirms the changes in the actual data set.. if we didn;t give that .. it wont change anything .. it wil just show u the temp output of the data..
```
